### Columns to Consider.

```python
test_data['Cibil Score'] = pd.to_numeric(test_data['Cibil Score'], errors='coerce')
```

```python
median_value = test_data['Cibil Score'].median()
test_data['Cibil Score'].fillna(median_value, inplace=True)
```

```python
test_data['EMPLOY CONSTITUTION'].value_counts()
```

```python
### Since Data is skewed.
median_value = test_data['TOTAL ASSET COST'].median()
test_data['TOTAL ASSET COST'].fillna(median_value, inplace=True)
```

```python
test_data['ADDRESS TYPE'].value_counts()
```

```python
# Values of interest
values_of_interest = ['Self/Spouse Owned', 'RESIDENCE','Parental']

# Filter column for specific values
filtered_df = test_data[test_data['ADDRESS TYPE'].isin(values_of_interest)]

# Sum of filtered values
filtered_sum = len(filtered_df['ADDRESS TYPE'])

total = filtered_sum
self_o = test_data['ADDRESS TYPE'].value_counts()[0]/total
re = test_data['ADDRESS TYPE'].value_counts()[1]/total
par = test_data['ADDRESS TYPE'].value_counts()[2]/total


np.random.seed(0)
fill_values = np.random.choice(['Self/Spouse Owned','RESIDENCE','Parental'], size=len(test_data[test_data['ADDRESS TYPE'].isna()]), p=[self_o, re,par])
test_data.loc[test_data['ADDRESS TYPE'].isna(), 'ADDRESS TYPE'] = fill_values
```

```python
test_data.info()
```

```python
### filling values.
total = test_data['EMPLOY CONSTITUTION'].notna().sum()
self_e = test_data['EMPLOY CONSTITUTION'].value_counts()[0]/total
sal = test_data['EMPLOY CONSTITUTION'].value_counts()[1]/total


np.random.seed(0)
fill_values = np.random.choice(['SELF-EMPLOYED', 'SALARIED'], size=len(test_data[test_data['EMPLOY CONSTITUTION'].isna()]), p=[self_e, sal])
test_data.loc[test_data['EMPLOY CONSTITUTION'].isna(), 'EMPLOY CONSTITUTION'] = fill_values
```
